```python
# importing Libraries
from scipy import stats
import numpy as np
```

# Data Provided:
-	The theoretical weekly operating cost model: W = $1,000 + $5X
-	Sample of 25 restaurants with a mean weekly cost of Rs. 3,050
-	Number of units produced in a week (X) follows a normal distribution with a mean (Î¼) of 600 units and a standard deviation (Ïƒ) of 25 units

```python
# According to given data
n=25
sam_mean_cost = 3050
pop_std = 25*5
pop_mean_cost = 1000 + 5*600
alpha = 0.05
```

# Assignment Tasks:
1. State the Hypotheses statement:

- H0 :Î¼=4000(TheÂ modelÂ isÂ correct)

- ð»a :ðœ‡>4000(ActualÂ costsÂ areÂ higher)

2. Calculate the Test Statistic:

-    Z = (x_bar - pop_mean) / std/sqrt(n)

```python
# Test statical
z_stat = (sam_mean_cost - pop_mean_cost) / (pop_std/ np.sqrt(n))
z_stat
```

```python
z_critical = stats.norm.ppf(1-alpha)
z_critical
```

# Compare: observed 

As Z_critcal > z_stat

Therefore, Fail to reject H0 hypothesis.	â€‹
There is no evidence supporting the claim that mean weekly cost is higher than 4000.

```python
if z_stat > z_critical:
    print("Decision: Reject H0 (evidence costs are higher).")
else:
    print("Decision: Fail to reject H0 (no evidence costs are higher).")
```
