# Chi-Square test 

- œá2=‚àë(O‚àíE)2/E
- Expected Frequency = (Row_total)*(col_total)/Grand Total
### Degrees of freedom:
- df = (row-1)(column-1)

Objective:
To use the Chi-Square test for independence to determine if there's a significant association between the type of smart home device purchased (Smart Thermostats vs. Smart Lights) and the customer satisfaction level.

### Data Provided:
The data is summarized in a contingency table showing the counts of customers in each satisfaction level for both types of devices:
- Satisfaction	Smart Thermostat	Smart Light	Total
- Very Satisfied	50	70	120
- Satisfied	80	100	180
- Neutral	60	90	150
- Unsatisfied	30	50	80
- Very Unsatisfied	20	50	70
- Total	240	360	600


```python
import numpy as np
import pandas as pd
from scipy.stats import chi2
```

Assignment Tasks:
1. State the Hypotheses:

H‚ÇÄ: Device type and customer satisfaction are independent.

H‚ÇÅ: Device type and customer satisfaction are associated.

2. Compute the Chi-Square Statistic:
- œá2=‚àë(O‚àíE)2/E

```python
observed = np.array([
    [50, 70],
    [80, 100],
    [60, 90],
    [30, 50],
    [20, 50]
])
```

```python
# Step 1: Compute chi-square manually
row_totals = observed.sum(axis=1).reshape(-1,1)
col_totals = observed.sum(axis=0).reshape(1,-1)
grand_total = observed.sum()

(row_totals,col_totals,grand_total)
```

```python
# Expected frequencies
expected = row_totals @ col_totals / grand_total
expected
```

```python
chi_square_stat = ((observed - expected)**2 / expected).sum()
chi_square_stat
```

#### 3. Determine the Critical Value:
Using the significance level (alpha) of 0.05 and the degrees of freedom (which is the number of categories minus 1)
- df = (row-1)(column-1)

```python
observed.shape
```

```python
dof = (observed.shape[0]-1) * (observed.shape[1]-1)
dof
```

```python
# ---- Step 5: Critical value & p-value ----
alpha = 0.05
chi_critical = chi2.ppf(1-alpha, dof)
p_value = 1 - chi2.cdf(chi_square_stat, dof)

(chi_critical,p_value)
```

### 4. Make a Decision:
Compare the Chi-Square statistic with the critical value to decide whether to reject the null hypothesis.

# Since, ùúí2=5.64<9.488, we fail to reject H‚ÇÄ.

# Final Conclusion

- There is no significant association between device type and customer satisfaction.
- Satisfaction does not depend on whether a customer bought a thermostat or a light.
