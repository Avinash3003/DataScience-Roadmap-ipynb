# Exercise 1: Prime Numbers

```python
# Importing libraries
import math

# Is_prime function returns boolean value as True if the given number is prime else false
def is_prime(n):
    if n <= 1:
        return False
    # O(sqrt(n)) approach
    for i in range(2,int(math.sqrt(n))+1):
        if n%i==0:
            return False
    return True


num = int(input('Enter prime number: '))
if is_prime(num):
    print(f"{num} is a Prime Number")
else:
    print(f"{num} is not a Prime Number")
```

# Exercise 4: Word counter

```python
s = input("Enter the Text: ")
l = s[:-1].split()
d={}

for i in l:
    if i not in d:
        d[i]=1
    else:
        d[i]+=1

for k,v in d.items():
    print(k,":",v)
```

# Exercise 5: Check for Palindrome

```python
def is_palindrome(s):
    return s==s[::-1]

# Removing Spaces, special Characters, Capitalized
s = input("Enter theh String: ")
s1=""
for i in s:
    if (i>='a' and i<='z') or (i>='A' and i<='Z'):
        s1+=i


if is_palindrome(s1.lower()):
    print(s1,"is a Palindrome")
else:
    print(s1,"is not a Palindrome")
```
